```python
data_path = "/var/Data/Cotton Imaging Datasets/Training Datasets/Dataset_01312022/Dataset80_20/DataSet1_20"
labels_path = "/var/Data/Cotton Imaging Datasets/Training Datasets/Dataset_01312022/Dataset80_20/mask_rcnn_R_101_FPN_3x/Model_5000/coco_instances_results.json"
# Create the dataset
dataset = fo.Dataset.from_dir(
    data_path=data_path,
    labels_path=labels_path,
    dataset_type=fo.types.COCODetectionDataset,
    label_field="predictions",
)

# Export images and ground truth labels to disk
dataset.export(
    dataset_type=fo.types.COCODetectionDataset,
    labels_path="/var/Data/Cotton Imaging Datasets/Training Datasets/Dataset_01312022/Dataset80_20/mask_rcnn_R_101_FPN_3x/Model_5000/FiftyOne Visualization/predictions.json",
    classes = ["openBoll", "closedBoll", "Flower", "Square"],
)

dataset.delete()
```

```python
# Now load ground truth labels into a new dataset
dataset = fo.Dataset.from_dir(
    dataset_dir="/var/Data/Cotton Imaging Datasets/Training Datasets/Dataset_01312022/Dataset80_20/mask_rcnn_R_101_FPN_3x/Model_5000/FiftyOne Visualization",
    dataset_type=fo.types.COCODetectionDataset,
    label_field="ground_truth",
)

# And add model predictions
fouc.add_coco_labels(
    dataset,
    "predictions",
    "/var/Data/Cotton Imaging Datasets/Training Datasets/Dataset_01312022/Dataset80_20/mask_rcnn_R_101_FPN_3x/Model_5000/FiftyOne Visualization/predictions.json",
    classes=["openBoll", "closedBoll", "Flower", "Square"],
)

session = fo.launch_app(dataset)
```
